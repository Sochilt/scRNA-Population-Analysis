```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Sample data
samples = ['S3216', 'S2792', 'S2553', 'C2495', 'C3006']
ITGAM_levels = [5.1, 4.5, 5.6, 4, 6.2]
CST3_levels = [0.5, 1.8, 2.0, 1.5, 0.7]
Th17_populations = [2.9, 4.7, 2.6, 4.9, 5.3]
Th2_populations = [48.9, 41.3, 31., 28.9, 49.6]
Th1_populations = [47.9, 42.2, 20.5, 28.6, 66.8]

# Create a DataFrame for the heatmap
data = {
    'ITGAM': ITGAM_levels,
    'CST3': CST3_levels,
    'Th17': Th17_populations,
    'Th2': Th2_populations,
    'Th1': Th1_populations
}
df = pd.DataFrame(data, index=samples)

# Plot bar charts
fig, axes = plt.subplots(5, 1, figsize=(10, 15))

axes[0].bar(samples, ITGAM_levels, color='blue')
axes[0].set_title('ITGAM Levels Across Samples')
axes[0].set_xlabel('Samples')
axes[0].set_ylabel('ITGAM Levels')

axes[1].bar(samples, CST3_levels, color='green')
axes[1].set_title('CST3 Levels Across Samples')
axes[1].set_xlabel('Samples')
axes[1].set_ylabel('CST3 Levels')

axes[2].bar(samples, Th17_populations, color='purple')
axes[2].set_title('Th17 Populations Across Samples')
axes[2].set_xlabel('Samples')
axes[2].set_ylabel('Th17 Populations')

axes[3].bar(samples, Th2_populations, color='red')
axes[3].set_title('Th2 Populations Across Samples')
axes[3].set_xlabel('Samples')
axes[3].set_ylabel('Th2 Populations')

axes[4].bar(samples, Th1_populations, color='black')
axes[4].set_title('Th1 Populations Across Samples')
axes[4].set_xlabel('Samples')
axes[4].set_ylabel('Th1 Populations')

plt.tight_layout()
plt.show()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.T, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap of ITGAM, CST3, & Helper T Across Samples')
plt.show()
```
